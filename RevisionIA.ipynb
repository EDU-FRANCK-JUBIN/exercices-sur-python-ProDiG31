```python
# Exo 01 : Sapin de Noël

def DrawBranch(b):
    pen.write(b * '^')
    pen.left(90)
    pen.forward(10)
    pen.right(90)
    pen.forward(5)
    if(b <= 1): return
    DrawBranch(b - 2)
        
nbBranch = int(input("Veuillez indiquer la taille de votre sapin de noël ? "))

pen = tu.Turtle()
pen.seth(0)
pen.up()
DrawBranch(nbBranch)
tu.exitonclick()
```

```python
#Exo 03 : Dessine cette scene 
import turtle as tu

def DrawLineAndRotate(lengh, angle):
    pen.forward(lengh)
    pen.left(angle)

def DrawEquiTriangle(lengh):
    DrawLineAndRotate(lengh, 120)
    DrawLineAndRotate(lengh, 120)
    DrawLineAndRotate(lengh, 120)

def DrawCircleTriangleFromCenter(lengh):
    for i in range(6):
        DrawEquiTriangle(lengh)
        pen.left(60)

def DrawBikeFrame(lengh):
    pen.seth(60)
    #Fourche
    DrawLineAndRotate(lengh,240)
    
    #Guidon
    pen.color('red')
    pen.seth(45)
    pen.forward(lengh/2)
    pen.backward(lengh/2)
    
    #Cadre
    pen.color('orange')
    pen.seth(-60)
    DrawEquiTriangle(lengh)
    pen.seth(0)
    pen.forward(lengh)
    
    #Selle
    pen.color('red')
    pen.seth(45)
    pen.forward(lengh/2)
    pen.right(45)
    pen.backward(lengh/2)
    pen.forward(lengh/2)
    pen.left(45)
    pen.backward(lengh/2)
    
    #Fourche arriere
    pen.color('orange')
    pen.seth(-120)
    DrawEquiTriangle(lengh)
    pen.seth(-60)
    pen.forward(lengh)
 

def DrawBike(lengh):
    pen.color('yellow')
    pen.left(30)
    DrawCircleTriangleFromCenter(stepSize)
    pen.color('orange')
    DrawBikeFrame(stepSize*2)
    pen.color('yellow')
    pen.left(30)
    DrawCircleTriangleFromCenter(stepSize)
    
    
pen = tu.Turtle()

stepSize = 40
pen.width(5)

pen.up()
pen.goto(stepSize,stepSize)
pen.down()

DrawBike(stepSize)

tu.exitonclick()
```
